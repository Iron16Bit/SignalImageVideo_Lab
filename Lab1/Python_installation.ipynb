# HOW TO INSTALL PYTHON, GIT, VSCODE and create a virtual env.

1. Download python https://www.python.org/downloads/ (the one of your OS - Windows, Mac, Linux)

2. download git (if you are on Windows). You can find the exe in the folder of this notebook

3. download vscode https://code.visualstudio.com/download

4. install python and be sure to MARK "ADD PYTHON 3.x to PATH" and "install launcher for all users" box

![img/Immagine 2023-10-18 133604.png](<attachment:Immagine 2023-10-18 133604.png>)

5. install git (for windows users). Be sure to mark these option. Then spam the "next" button

![img/Immagine 2023-10-18 134105.png](<attachment:Immagine 2023-10-18 134105.png>)

![img/Immagine 2023-10-18 134127.png](<attachment:Immagine 2023-10-18 134127.png>)

6. install vscode

7. Open VsCode

8. search for the "Terminal" tab

![Screenshot from 2023-10-24 14-25-12.png](<attachment:Screenshot from 2023-10-24 14-25-12.png>)

9. Open Terminal (click on "New Terminal")

10. type at the same time the keys "Ctrl+Shift+p" , it will open a floating window as the one in the picture below. Then type "Terminal: Select Default Profile" and press enter.

![Screenshot from 2023-10-24 14-37-03.png](<attachment:Screenshot from 2023-10-24 14-37-03.png>)

11. Select bash or gitbash. if you are on Mac, you can keep zsh.

![Screenshot from 2023-10-24 14-40-08.png](<attachment:Screenshot from 2023-10-24 14-40-08.png>)

12. In the terminal type ''python --version'' and press enter, if it does not print python 3.x.x, type ''python3 --version'' 

13. In the terminal type ''pip --version'' and press enter

14. In the terminal type ''pip install virtualenv'' and press enter

15. In the terminal type ''python -m venv myenv_lab1'' and press enter, if in 12. you had to run ''python3 --version'', then type ''python3 -m venv myenv_lab1'' . This step will create a virtual environment where you will store all the libraries to run your code. The main pros of having a virtual env are: a. if it broke it does not broke your PC, b. if it broke, you can delete it and create it again, c. if you create a new environment it does not install from scratch the same libraries (conda does that) but point to libraries already installed in precedent environment. **NOTE: Always install the virtual env in the forlder where you have your code!!!**

![Screenshot from 2023-10-24 14-48-30.png](<attachment:Screenshot from 2023-10-24 14-48-30.png>)

16. Activate your environment! For Windows users: ''source myenv_lab1/Scripts/activate'' ; all other OS ''source myenv_lab1/bin/activate''

17. install libraries typing in the terminal ''pip install -r requisites.txt''

18. Last steps, click on ''Select Kernel'' tab 

![Screenshot from 2023-10-24 14-57-48.png](<attachment:Screenshot from 2023-10-24 14-57-48.png>)

19. Click on ''Python Environments''

![Screenshot from 2023-10-24 14-58-19.png](<attachment:Screenshot from 2023-10-24 14-58-19.png>)

20. select your environment ''myenv_lab1"
