## Objective: 

The objective of this exercise is to help beginners get familiar with NumPy, scipy, scikit-learn, and OpenCV while performing basic image processing and analysis tasks.

```python
#read image

import cv2

# Load the image
image = cv2.imread('imgs_lab/sample_image.png')
```

Display image

```python
import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

Convert to grayscale and display it

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

Display every color channel [r,g,b]

```python
b,g,r=cv2.split(image)

plt.subplot(223),plt.imshow(b, cmap='gray'),plt.title('Blue')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(g, cmap='gray'),plt.title('Green')
plt.xticks([]), plt.yticks([])
plt.subplot(221),plt.imshow(r, cmap='gray'),plt.title('Red')
plt.xticks([]), plt.yticks([])
plt.show()
```

Flip the image

```python
flipped_image = cv2.flip(image, 0)
plt.imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
```

Crop the image

```python
cropped_image = image[200:300, 130:300]
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
```

Change color of a box in the image

```python
import numpy as np
#First create an image the same size as our input
blank_image = np.zeros((image.shape), np.uint8)
blank_image[0:400,0:400,1]=100; #give it a green square
new_image=cv2.add(blank_image,image) # add the two images together
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
```

## The Extra-Mile! 

Using the image rocky.jpg generate rocky_chess_pattern.png

```python
image_path = 'imgs_lab/rocky.jpg'  
patch_size = 128  
input_image=cv2.imread(image_path)

#Height, Width, Channel
h,w,c = input_image.shape

boxes_h = h//patch_size
boxes_w = w//patch_size

color = 0
for i in range(boxes_h+1):
    for j in range(boxes_w+1):
        blank_image = np.zeros((input_image.shape), np.uint8)
        if color%2 == 0:
            blank_image[i*patch_size:i*patch_size+patch_size,j*patch_size:j*patch_size+patch_size,1]=50
        else:
            blank_image[i*patch_size:i*patch_size+patch_size,j*patch_size:j*patch_size+patch_size,2]=50
        input_image=cv2.add(blank_image, input_image)
        color+=1

plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
```
