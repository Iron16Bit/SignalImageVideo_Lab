# Exercise 2.2: Basic Image Processing and Analysis

## Objective: 

Learn how to apply various filters to an image using OpenCV, NumPy, and Matplotlib.

```python
#read image

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('imgs_lab/lena_color.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

Implement functions to apply various filters (gaussian blur, median blur and sobel edge detection) to the image

- https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html

- https://www.geeksforgeeks.org/python-program-to-detect-the-edges-of-an-image-using-opencv-sobel-edge-detection/


```python
gauss_blur = cv2.GaussianBlur(image,(5,5),9)

median_blur = cv2.medianBlur(image,9)

sobelimage=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
sobel_edges = cv2.Sobel(sobelimage, cv2.CV_64F, 1, 1, ksize=9)
```

Display the original image and the filtered images using Matplotlib

```python
plt.subplot(223),plt.imshow(cv2.cvtColor(gauss_blur, cv2.COLOR_BGR2RGB)),plt.title('Gaussian')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB)),plt.title('Median')
plt.xticks([]), plt.yticks([])
plt.subplot(221),plt.imshow(sobel_edges, cmap='gray'),plt.title('Sobel')
plt.xticks([]), plt.yticks([])
plt.show()
```
